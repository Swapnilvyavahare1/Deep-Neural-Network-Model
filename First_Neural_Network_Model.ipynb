```python
from tensorflow import keras
```

```python
df= pd.read_csv('Boston.csv')
```

```python
X= df.iloc[:,1:-1]
y= df.iloc[:,-1]
```

```python
y
```

```python
# Split the data into training and test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state= 1)
```

**DNN**

```python
# Sequntial API
dnn= tf.keras.models.Sequential()
```

```python
# Adding first hidden layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding second hidden layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding output layer
dnn.add(tf.keras.layers.Dense(units=1))
```

```python
# Compiling the network with optimization method
dnn.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model with input dataset
dnn.fit(x_train, y_train, batch_size= 32, epochs= 100)
```

```python
dnn.fit(x_train, y_train, batch_size= 32, epochs= 200)
```

```python
# Prediction
y_pred= dnn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
